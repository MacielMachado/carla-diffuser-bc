```python
import os
import numpy as np

# Caminho para a pasta com os arquivos .txt
pasta_txt = 'txt_bc/txt_bc'

# Lista de classes
classes = [
    'gail_experts_nroutes1_neps1_ce06_ep_1_',
    'gail_experts_nroutes1_neps1_ce06_ep_20_',
    'gail_experts_nroutes1_neps1_ce06_ep_30_',
    'gail_experts_nroutes1_neps1_ce06_ep_40_',
    'gail_experts_nroutes1_neps1_ce06_ep_50_',
    'gail_experts_nroutes1_neps1_ce06_ep_60_',
    'gail_experts_nroutes1_neps1_ce06_ep_70_',
    'gail_experts_nroutes1_neps1_ce06_ep_80_',
    'gail_experts_nroutes1_neps1_ce06_ep_90_',
    'gail_experts_nroutes1_neps1_ce06_ep_100',
    'gail_experts_nroutes1_neps1_ce06_ep_120',
    'gail_experts_nroutes1_neps1_ce06_ep_150',
    'gail_experts_nroutes1_neps1_ce06_ep_200',
    'gail_experts_nroutes1_neps1_ce06_ep_250',
]

# Dicionário para armazenar os contadores de comprimentos e distâncias por classe
contadores_comprimentos = {classe: {'100-149': 0, '150-199': 0, '200-249': 0, '250-299': 0, '300': 0} for classe in classes}
distancias_totais = {classe: 0.0 for classe in classes}
quantidades = {classe: 0 for classe in classes}

# Função para calcular a distância Euclidiana entre dois pontos 3D
def distancia_euclidiana(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

# Percorrer todos os arquivos na pasta
for filename in os.listdir(pasta_txt):
    if filename.endswith('.txt'):
        filepath = os.path.join(pasta_txt, filename)
        # Determinar a classe do arquivo
        for classe in classes:
            if classe in filename and filename not in straight_files:
                with open(filepath, 'r') as file:
                    # Ler os dados do arquivo
                    linhas = file.readlines()
                    comprimento = len(linhas)
                    
                    # Atualizar o contador de comprimentos
                    if 100 <= comprimento <= 149:
                        contadores_comprimentos[classe]['100-149'] += 1
                    elif 150 <= comprimento <= 199:
                        contadores_comprimentos[classe]['150-199'] += 1
                    elif 200 <= comprimento <= 249:
                        contadores_comprimentos[classe]['200-249'] += 1
                    elif 250 <= comprimento <= 299:
                        contadores_comprimentos[classe]['250-299'] += 1
                    elif comprimento == 300:
                        contadores_comprimentos[classe]['300'] += 1

                    # Calcular a distância total percorrida
                    coordenadas = [list(map(float, linha.split())) for linha in linhas]
                    distancia_total = sum(distancia_euclidiana(coordenadas[i], coordenadas[i+1]) for i in range(len(coordenadas) - 1))
                    distancias_totais[classe] += distancia_total
                    quantidades[classe] += 1
                
                break

# Calcular a distância média por classe
distancias_medias = {classe: (distancias_totais[classe] / quantidades[classe]) if quantidades[classe] > 0 else 0 for classe in classes}

# Exibir os resultados
for classe, contadores in contadores_comprimentos.items():
    print(f'Classe {classe}:')
    for comprimento, contador in contadores.items():
        print(f'  {comprimento}: {contador}')
```
