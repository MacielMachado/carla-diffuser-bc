```python
import os
from statistics import mean
import matplotlib.pyplot as plt

# Caminho para a pasta com os arquivos .txt
pasta_txt = 'txt/txt'

# Lista de classes
classes = [
    'ep_40',
    'ep_50',
    'ep_60',
    'ep_70',
    'ep_80',
    'ep_90',
    'ep_100',
    'ep_120',
    'ep_150',
]

# Dicionário para armazenar os tamanhos dos arquivos por classe
tamanhos_classes = {classe: [] for classe in classes}

# Percorrer todos os arquivos na pasta
for filename in os.listdir(pasta_txt):
    if filename.endswith('.txt'):
        filepath = os.path.join(pasta_txt, filename)
        # Determinar a classe do arquivo
        for classe in classes:
            if classe in filename:
                with open(filepath, 'r') as file:
                    # Carregar a lista de coordenadas
                    linhas = file.readlines()
                    # Adicionar o tamanho da lista à classe correspondente
                    tamanhos_classes[classe].append(len(linhas))
                    # print(linhas)
                break

# Calcular a média de tamanho de cada classe
medias_tamanhos = {classe: mean(tamanhos) for classe, tamanhos in tamanhos_classes.items() if len(tamanhos) > 0}
min_tamanhos = {classe: min(tamanhos) for classe, tamanhos in tamanhos_classes.items() if len(tamanhos) > 0}
max_tamanhos = {classe: max(tamanhos) for classe, tamanhos in tamanhos_classes.items() if len(tamanhos) > 0}

classes_lista = list(medias_tamanhos.keys())
medias_lista = list(medias_tamanhos.values())
min_lista = [medias_tamanhos[classe] - min_tamanhos[classe] for classe in classes_lista]
max_lista = [max_tamanhos[classe] - medias_tamanhos[classe] for classe in classes_lista]

# Gerar o histograma
plt.figure(figsize=(12, 6))
# plt.bar(medias_tamanhos.keys(), medias_tamanhos.values(), color='b', align='center')

plt.bar(classes_lista, medias_lista, color='b', align='center', yerr=[min_lista, max_lista], capsize=5, error_kw={'ecolor': 'red', 'capsize': 5}
)

plt.xlabel('Classes')
plt.ylabel('Média de Tamanho')
plt.title('Histograma da Média de Tamanho das Classes')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Caminho para a pasta com os arquivos .txt
pasta_txt = 'txt/txt'

# Lista de classes
classes = [
    'gail_experts_nroutes1_neps1_8290_ep_40',
    'gail_experts_nroutes1_neps1_8290_ep_80',
    'gail_experts_nroutes1_neps1_8290_ep_100',
]

# Dicionário para armazenar os contadores de comprimentos e distâncias por classe
contadores_comprimentos = {classe: {'100-149': 0, '150-199': 0, '200-249': 0, '250-299': 0, '300': 0} for classe in classes}
distancias_totais = {classe: 0.0 for classe in classes}
quantidades = {classe: 0 for classe in classes}

# Função para calcular a distância Euclidiana entre dois pontos 3D
def distancia_euclidiana(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

# Percorrer todos os arquivos na pasta
for filename in os.listdir(pasta_txt):
    if filename.endswith('.txt'):
        filepath = os.path.join(pasta_txt, filename)
        # Determinar a classe do arquivo
        for classe in classes:
            if classe in filename:
                with open(filepath, 'r') as file:
                    # Ler os dados do arquivo
                    linhas = file.readlines()
                    comprimento = len(linhas)
                    
                    # Atualizar o contador de comprimentos
                    if 100 <= comprimento <= 149:
                        contadores_comprimentos[classe]['100-149'] += 1
                    elif 150 <= comprimento <= 199:
                        contadores_comprimentos[classe]['150-199'] += 1
                    elif 200 <= comprimento <= 249:
                        contadores_comprimentos[classe]['200-249'] += 1
                    elif 250 <= comprimento <= 299:
                        contadores_comprimentos[classe]['250-299'] += 1
                    elif comprimento == 300:
                        contadores_comprimentos[classe]['300'] += 1

                    # Calcular a distância total percorrida
                    coordenadas = [list(map(float, linha.split())) for linha in linhas]
                    distancia_total = sum(distancia_euclidiana(coordenadas[i], coordenadas[i+1]) for i in range(len(coordenadas) - 1))
                    distancias_totais[classe] += distancia_total
                    quantidades[classe] += 1
                
                break

# Calcular a distância média por classe
distancias_medias = {classe: (distancias_totais[classe] / quantidades[classe]) if quantidades[classe] > 0 else 0 for classe in classes}

# Exibir os resultados
for classe, contadores in contadores_comprimentos.items():
    print(f'Classe {classe}:')
    for comprimento, contador in contadores.items():
        print(f'  {comprimento}: {contador}')
    print(f'  Distância Média: {distancias_medias[classe]:.2f}')

# Calcular as porcentagens e preparar os dados para o histograma
faixas = ['100-149', '150-199', '200-249', '250-299', '300']
dados = {classe: [contadores_comprimentos[classe][faixa] / sum(contadores_comprimentos[classe].values()) * 100 for faixa in faixas] for classe in classes}

# Configuração do histograma
bar_width = 0.15
x = np.arange(len(classes))

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 8))

# Cores para as faixas
cores = ['blue', 'orange', 'green', 'red', 'purple']

# Plotar os dados para cada faixa
for i, faixa in enumerate(faixas):
    valores = [dados[classe][i] for classe in classes]
    ax.bar(x + i * bar_width, valores, bar_width, label=faixa, color=cores[i])

# Adicionar detalhes ao gráfico
ax.set_xlabel('Epochs')
ax.set_ylabel('Duration (%)')
ax.set_title('Distribuição de Comprimentos por Classe')
ax.set_xticks(x + bar_width * (len(faixas) - 1) / 2)
ax.set_xticklabels(['40', '80', '100'])
ax.legend()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
